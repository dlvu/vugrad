# Question 5

```python
a = vg.TensorNode(np.array([[1,2],[3,4]]))
b = vg.TensorNode(np.array([[5,6],[7,8]]))

c = a+b

print(f"c.value : \n{c.value}")
print(f"\nc.source : \n{c.source}")
print(f"\nc.source.inputs : \n{c.source.inputs}")
print(f"\nc.source.inputs[0] : \n{c.source.inputs[0]}")
print(f"\nc.source.inputs[0].value : \n{c.source.inputs[0].value}")
print(f"\na.grad : \n{a.grad}")
print(f"\nc.grad : \n{c.grad}")
```

# Question 8

```python
# Try of Expand operation 
X = vg.TensorNode(np.array([[1,2,3,4,5]]))

Y = vg.Expand.do_forward(X, dim=0, repeats=5)

print(f"TensorNode X value:\n {X.value}")
print(f"\nTensorNode Y=extend(X) value:\n{Y.value}")

Y_nabla = Y.value
print(f"\nSimulation of backward of X: \n{Y_nabla.sum(axis=0, keepdims=True)}")
```
